## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
```

```python
# Step 1: Rename columns for consistency
users.rename(columns={'code': 'userCode', 'name': 'userName'}, inplace=True)
hotels.rename(columns={'name': 'hotelName'}, inplace=True)

# Step 2: Convert date columns to datetime
hotels['date'] = pd.to_datetime(hotels['date'], format='%m/%d/%Y')
flights['date'] = pd.to_datetime(flights['date'], format='%m/%d/%Y')

# Step 3: Check for missing values
missing_hotels = hotels.isnull().sum()
missing_users = users.isnull().sum()
missing_flights = flights.isnull().sum()

# Step 4: Drop duplicates if any
hotels.drop_duplicates(inplace=True)
users.drop_duplicates(inplace=True)
flights.drop_duplicates(inplace=True)

# Step 5: Ensure correct data types
# Already handled for dates and basic types are fine

# Output cleaned info
hotels_display = hotels.head(10)
users_display = users.head(10)
flights_display = flights.head(10)

missing_hotels, missing_users, missing_flights
```

#### Chart - 9

```python
# Chart - 9 visualization code
```

```python
top_agencies = flights['agency'].value_counts().nlargest(5).index
filtered_flights = flights[flights['agency'].isin(top_agencies)]
flight_type_by_agency = pd.crosstab(filtered_flights['agency'], filtered_flights['flightType'])

flight_type_by_agency.plot(kind='bar', stacked=True, figsize=(10, 6), title='Flight Type Count by Top 5 Agencies')
plt.xlabel('Agency')
plt.ylabel('Number of Flights')
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

To see the flight seat type by top agencies.

##### 2. What is/are the insight(s) found from the chart?

FlyingDrops only offers a higher segment of seats, hence low business as seen earlier.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

FlyingDrops can consider investing in cheaper seat segments and not just the luxury segment, to increase their customer demographic.
